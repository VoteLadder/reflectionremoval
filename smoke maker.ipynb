```python
import os
import cv2
import numpy as np
from scipy.ndimage import gaussian_filter
import random
import logging
from tqdm import tqdm
from concurrent.futures import ProcessPoolExecutor, as_completed

# Optional imports for GPU and visualization will be handled within functions
try:
    import torch
    import torch.nn.functional as F
except ImportError:
    torch = None

# Set up logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s [%(levelname)s] %(message)s")

# === GLOBAL PARAMETERS FOR SMOKE EFFECT ===
# Smoke generation parameters
NUM_CLOUDS = 35                           # Even more clouds for better coverage
MIN_RADIUS_FACTOR = 0.1                   # Smaller min radius for more varied effects
MAX_RADIUS_FACTOR = 0.8                   # Larger max radius for bigger coverage
PERLIN_SCALE_MIN = 8                      # Lower for even larger texture patterns
PERLIN_SCALE_MAX = 150                    # Higher max scale for more extreme variation
PERLIN_OCTAVES = 9                        # More detail layers for complexity
PERLIN_PERSISTENCE = 0.7                  # Higher persistence for more contrast

# Smoke appearance parameters
INTENSITY_RANGE = (1.2, 2.5)              # Significantly higher for near-complete opacity
BLUR_RANGE = (8, 80)                      # Adjusted for sharper patches
COLOR_RANGE = ((60, 150), (60, 150), (60, 150))  # Even darker smoke color range

# Processing parameters
NUM_VARIATIONS = 1                       # Number of smoke variations per image
TOTAL_IMAGES = None                      # Set to None to process all images
NUM_WORKERS = 4                          # Number of parallel workers
USE_GPU = True                           # Set True to use GPU if available

# Input/Output directories
CLEAR_DIR = "output_3_18/classified_blurry"  # Source directory with clear images
OUTPUT_DIR = "output_3_18/smokeV2"           # Output directory for processed images

# === CPU Functions ===

def generate_perlin_noise(shape, scale=None, octaves=PERLIN_OCTAVES, persistence=None):
    """Generate a multi-octave Perlin noise pattern using NumPy for CPU."""
    if scale is None:
        scale = random.uniform(PERLIN_SCALE_MIN, PERLIN_SCALE_MAX)
    if persistence is None:
        persistence = random.uniform(0.4, 0.8)  # Random persistence for more variation
        
    def generate_octave(shape, scale):
        x = np.linspace(0, 1, shape[1]) * scale
        y = np.linspace(0, 1, shape[0]) * scale
        x_idx, y_idx = np.meshgrid(x, y)
        seed = random.randint(0, 1000)
        np.random.seed(seed)
        angles = np.random.rand(4) * 2 * np.pi
        freqs = np.random.rand(4) * 10 + 0.1
        noise = np.zeros(shape)
        for i in range(4):
            noise += np.sin(x_idx * np.cos(angles[i]) * freqs[i] + 
                            y_idx * np.sin(angles[i]) * freqs[i])
        noise = (noise - noise.min()) / (noise.max() - noise.min())
        return noise
        
    noise = np.zeros(shape)
    max_val = 0
    for i in range(octaves):
        octave_scale = scale * (2 ** i)
        weight = persistence ** i  # Use persistence instead of fixed ratio
        max_val += weight
        noise += generate_octave(shape, octave_scale) * weight
    noise = noise / max_val
    return noise

def create_smoke_mask(shape, num_clouds=NUM_CLOUDS, min_radius=None, max_radius=None):
    """Create a mask with random smoke-like areas using NumPy for CPU."""
    # Add randomization to cloud density - increased chance of very dense clusters
    cloud_density = random.choice([1, 1, 2, 3, 4])  # More frequently have denser clusters
    num_clouds = num_clouds * cloud_density
    
    if min_radius is None:
        min_radius = int(min(shape) * MIN_RADIUS_FACTOR)
    if max_radius is None:
        max_radius = int(min(shape) * MAX_RADIUS_FACTOR)
        
    mask = np.zeros(shape, dtype=np.float32)
    
    # Create random clusters of clouds
    num_clusters = random.randint(1, 4)
    for _ in range(num_clusters):
        cluster_center_x = random.randint(0, shape[1]-1)
        cluster_center_y = random.randint(0, shape[0]-1)
        cluster_radius = random.randint(max_radius//2, max_radius)
        clouds_in_cluster = num_clouds // num_clusters
        
        for _ in range(clouds_in_cluster):
            # Position within cluster area
            angle = random.uniform(0, 2 * np.pi)
            dist = random.uniform(0, cluster_radius)
            center_x = int(cluster_center_x + dist * np.cos(angle))
            center_y = int(cluster_center_y + dist * np.sin(angle))
            
            # Clamp to image boundaries
            center_x = max(0, min(center_x, shape[1]-1))
            center_y = max(0, min(center_y, shape[0]-1))
            
            radius = random.randint(min_radius, max_radius)
            y, x = np.ogrid[-center_y:shape[0]-center_y, -center_x:shape[1]-center_x]
            dist_from_center = np.sqrt(x*x + y*y)
            
            # Use different falloff functions randomly
            if random.random() > 0.7:
                # Sharper edges
                falloff = 1.5
                mask_cloud = np.clip(1.0 - (dist_from_center / radius) ** falloff, 0, 1)
            else:
                # Gaussian falloff
                mask_cloud = np.exp(-(dist_from_center**2) / (2. * (radius**2)))
                
            mask = np.maximum(mask, mask_cloud)
    
    return mask

# === GPU-Compatible Functions ===

def generate_perlin_noise_torch(shape, scale=None, octaves=PERLIN_OCTAVES, persistence=None, device='cpu'):
    """Generate Perlin noise using PyTorch for GPU acceleration."""
    if torch is None:
        raise ImportError("PyTorch is required for GPU processing.")
        
    if scale is None:
        scale = random.uniform(PERLIN_SCALE_MIN, PERLIN_SCALE_MAX)
    if persistence is None:
        persistence = random.uniform(0.4, 0.8)  # Random persistence for more variation
        
    H, W = shape
    noise = torch.zeros((H, W), device=device)
    max_val = 0.0
    for i in range(octaves):
        octave_scale = scale * (2 ** i)
        weight = persistence ** i  # Use persistence instead of fixed ratio
        max_val += weight
        x = torch.linspace(0, 1, W, device=device) * octave_scale
        y = torch.linspace(0, 1, H, device=device) * octave_scale
        x_idx, y_idx = torch.meshgrid(x, y, indexing="xy")
        angles = torch.rand(4, device=device) * 2 * np.pi
        freqs = torch.rand(4, device=device) * 10 + 0.1
        noise_octave = torch.zeros((H, W), device=device)
        for j in range(4):
            noise_octave += torch.sin(x_idx * torch.cos(angles[j]) * freqs[j] + 
                                     y_idx * torch.sin(angles[j]) * freqs[j])
        noise_octave = (noise_octave - noise_octave.min()) / (noise_octave.max() - noise_octave.min() + 1e-8)
        noise += noise_octave * weight
    noise = noise / (max_val + 1e-8)
    return noise

def create_smoke_mask_torch(shape, num_clouds=NUM_CLOUDS, min_radius=None, max_radius=None, device='cpu'):
    """Create a smoke mask using PyTorch for GPU acceleration."""
    if torch is None:
        raise ImportError("PyTorch is required for GPU processing.")
    
    # Add randomization to cloud density - increased chance of very dense clusters
    cloud_density = random.choice([1, 1, 2, 3, 4])  # More frequently have denser clusters
    num_clouds = num_clouds * cloud_density
    
    H, W = shape
    if min_radius is None:
        min_radius = int(min(H, W) * MIN_RADIUS_FACTOR)
    if max_radius is None:
        max_radius = int(min(H, W) * MAX_RADIUS_FACTOR)
        
    mask = torch.zeros((H, W), device=device)
    
    # Create random clusters of clouds
    num_clusters = random.randint(1, 4)
    for _ in range(num_clusters):
        cluster_center_x = random.randint(0, W-1)
        cluster_center_y = random.randint(0, H-1)
        cluster_radius = random.randint(max_radius//2, max_radius)
        clouds_in_cluster = num_clouds // num_clusters
        
        for _ in range(clouds_in_cluster):
            # Position within cluster area
            angle = random.uniform(0, 2 * np.pi)
            dist = random.uniform(0, cluster_radius)
            center_x = int(cluster_center_x + dist * np.cos(angle))
            center_y = int(cluster_center_y + dist * np.sin(angle))
            
            # Clamp to image boundaries
            center_x = max(0, min(center_x, W-1))
            center_y = max(0, min(center_y, H-1))
            
            radius = random.randint(min_radius, max_radius)
            y = torch.arange(0, H, device=device) - center_y
            x = torch.arange(0, W, device=device) - center_x
            y = y.unsqueeze(1)  # shape (H, 1)
            x = x.unsqueeze(0)  # shape (1, W)
            dist = torch.sqrt(x**2 + y**2)
            
            # Use different falloff functions randomly
            if random.random() > 0.7:
                # Sharper edges
                falloff = 1.5
                mask_cloud = torch.clamp(1.0 - (dist / radius) ** falloff, 0, 1)
            else:
                # Gaussian falloff
                mask_cloud = torch.exp(- (dist**2) / (2 * (radius**2)))
                
            mask = torch.maximum(mask, mask_cloud)
    
    return mask

def gaussian_blur_torch(img, sigma, device='cpu'):
    """Apply Gaussian blur using PyTorch for GPU acceleration."""
    if torch is None:
        raise ImportError("PyTorch is required for GPU processing.")
    kernel_size = int(2 * round(3 * sigma) + 1)
    if kernel_size % 2 == 0:
        kernel_size += 1
    x = torch.arange(kernel_size, device=device) - kernel_size // 2
    gaussian_kernel_1d = torch.exp(- (x**2) / (2 * sigma**2))
    gaussian_kernel_1d = gaussian_kernel_1d / gaussian_kernel_1d.sum()
    gaussian_kernel_2d = torch.outer(gaussian_kernel_1d, gaussian_kernel_1d)
    gaussian_kernel_2d = gaussian_kernel_2d.unsqueeze(0).unsqueeze(0)  # (1, 1, k, k)
    channels = img.shape[1]
    gaussian_kernel_2d = gaussian_kernel_2d.repeat(channels, 1, 1, 1)
    padding = kernel_size // 2
    blurred = F.conv2d(img, weight=gaussian_kernel_2d, groups=channels, padding=padding)
    return blurred

# === Smoke Generation Functions ===

def add_smoke_to_image(image, intensity_range=INTENSITY_RANGE, blur_range=BLUR_RANGE, 
                      color_range=COLOR_RANGE, use_gpu=USE_GPU, num_variations=NUM_VARIATIONS):
    """
    Add smoke effect to an image, generating multiple variations.
    Uses GPU acceleration if use_gpu=True, otherwise processes on CPU.
    """
    smoked_images = []
    smoke_masks = []
    if use_gpu and torch is not None and torch.cuda.is_available():
        device = torch.device("cuda")
        image_t = torch.from_numpy(image).float().to(device).permute(2, 0, 1)  # (3, H, W)
        H, W = image.shape[:2]
        for _ in range(num_variations):
            # Generate base noise and mask
            persistence = random.uniform(0.4, 0.8)
            noise = generate_perlin_noise_torch((H, W), device=device, persistence=persistence)
            mask = create_smoke_mask_torch((H, W), device=device)
            combined = noise * mask
            
            # Apply varied blurring
            blur_sigma = random.uniform(blur_range[0], blur_range[1])
            smoke = gaussian_blur_torch(combined.unsqueeze(0).unsqueeze(0), sigma=blur_sigma, device=device).squeeze()
            
            # Add more extreme dark patches for patchy opacity
            if random.random() > 0.2:  # 80% chance to add dark patches
                # Create primary dark patches
                dark_patches = create_smoke_mask_torch((H, W), num_clouds=random.randint(5, 12), 
                                                    min_radius=int(min(H, W) * 0.05),
                                                    max_radius=int(min(H, W) * 0.25),
                                                    device=device)
                dark_intensity = random.uniform(0.7, 0.95)  # Much darker patches
                smoke = smoke * (1 - dark_patches * dark_intensity)
                
                # Add some extreme opacity patches
                if random.random() > 0.4:  # 60% chance for additional extreme patches
                    opaque_patches = create_smoke_mask_torch((H, W), num_clouds=random.randint(2, 6), 
                                                        min_radius=int(min(H, W) * 0.03),
                                                        max_radius=int(min(H, W) * 0.15),
                                                        device=device)
                    # Apply perlin noise to make the opacity uneven within the patches
                    opacity_noise = generate_perlin_noise_torch((H, W), scale=random.uniform(3, 8), 
                                                           octaves=2, device=device)
                    opaque_patches = opaque_patches * (opacity_noise * 0.3 + 0.7)  # Make noise less dominant
                    opaque_intensity = random.uniform(0.9, 1.0)  # Near complete opacity
                    smoke = smoke * (1 - opaque_patches * opaque_intensity)
            
            # Apply intensity with random variations
            base_intensity = random.uniform(intensity_range[0], intensity_range[1])
            # Create intensity variation mask
            intensity_var = generate_perlin_noise_torch((H, W), scale=random.uniform(5, 15), octaves=3, device=device)
            intensity_var = intensity_var * 0.4 + 0.8  # Range from 0.8 to 1.2
            intensity = base_intensity * intensity_var
            smoke = smoke * intensity
            
            # Create smoke with varied colors
            color_vals = [random.randint(color_range[c][0], color_range[c][1]) for c in range(3)]
            # Add color variation
            color_var = [random.uniform(0.8, 1.2) for _ in range(3)]
            overlay = torch.stack([
                smoke * color_vals[c] * color_var[c] for c in range(3)
            ], dim=0)  # (3, H, W)
            
            # Apply the smoke overlay with additive blending
            smoked = image_t + overlay
            smoked = torch.clamp(smoked, 0, 255)
            
            # Random color adjustments
            if random.random() > 0.5:
                hue_shift = torch.tensor([
                    random.uniform(0.9, 1.1),
                    random.uniform(0.9, 1.1),
                    random.uniform(0.9, 1.1)
                ], device=device).view(3, 1, 1)
                smoked = torch.clamp(smoked * hue_shift, 0, 255)
            
            smoked_images.append(smoked.permute(1, 2, 0).cpu().numpy().astype(np.uint8))
            smoke_masks.append(smoke.cpu().numpy())
            
            # Check GPU memory usage
            mem_used = torch.cuda.memory_allocated(device) / 1024**3  # GB
            mem_total = torch.cuda.get_device_properties(device).total_memory / 1024**3  # GB
            if mem_used / mem_total > 0.8:
                logging.warning(f"GPU memory usage high: {mem_used:.2f} GB / {mem_total:.2f} GB")
    else:
        for _ in range(num_variations):
            result = image.copy()
            
            # Generate base noise and mask with more variation
            persistence = random.uniform(0.4, 0.8)
            noise = generate_perlin_noise(image.shape[:2], persistence=persistence)
            mask = create_smoke_mask(image.shape[:2])
            combined = noise * mask
            
            # Apply varied blurring
            blur_sigma = random.uniform(blur_range[0], blur_range[1])
            smoke = gaussian_filter(combined, sigma=blur_sigma)
            
            # Add more extreme dark patches for patchy opacity
            if random.random() > 0.2:  # 80% chance to add dark patches
                # Create primary dark patches
                dark_patches = create_smoke_mask(image.shape[:2], num_clouds=random.randint(5, 12),
                                              min_radius=int(min(image.shape[:2]) * 0.05),
                                              max_radius=int(min(image.shape[:2]) * 0.25))
                dark_intensity = random.uniform(0.7, 0.95)  # Much darker patches
                smoke = smoke * (1 - dark_patches * dark_intensity)
                
                # Add some extreme opacity patches
                if random.random() > 0.4:  # 60% chance for additional extreme patches
                    opaque_patches = create_smoke_mask(image.shape[:2], num_clouds=random.randint(2, 6),
                                                    min_radius=int(min(image.shape[:2]) * 0.03),
                                                    max_radius=int(min(image.shape[:2]) * 0.15))
                    # Apply perlin noise to make the opacity uneven within the patches
                    opacity_noise = generate_perlin_noise(image.shape[:2], scale=random.uniform(3, 8), octaves=2)
                    opaque_patches = opaque_patches * (opacity_noise * 0.3 + 0.7)  # Make noise less dominant
                    opaque_intensity = random.uniform(0.9, 1.0)  # Near complete opacity
                    smoke = smoke * (1 - opaque_patches * opaque_intensity)
            
            # Apply intensity with random variations
            base_intensity = random.uniform(intensity_range[0], intensity_range[1])
            # Create intensity variation mask
            intensity_var = generate_perlin_noise(image.shape[:2], scale=random.uniform(5, 15), octaves=3)
            intensity_var = intensity_var * 0.4 + 0.8  # Range from 0.8 to 1.2
            intensity = base_intensity * intensity_var
            smoke = smoke * intensity
            
            # Create smoke with varied colors
            smoke_overlay = np.zeros_like(image)
            color_vals = [random.randint(color_range[c][0], color_range[c][1]) for c in range(3)]
            # Add color variation
            color_var = [random.uniform(0.8, 1.2) for _ in range(3)]
            for c in range(3):
                smoke_overlay[:, :, c] = smoke * color_vals[c] * color_var[c]
            
            # Apply the smoke overlay with additive blending
            result = cv2.addWeighted(result, 1.0, smoke_overlay.astype(np.uint8), 1.0, 0)
            
            # Random color adjustments
            if random.random() > 0.5:
                hue_shifts = [random.uniform(0.9, 1.1) for _ in range(3)]
                for c in range(3):
                    result[:, :, c] = np.clip(result[:, :, c] * hue_shifts[c], 0, 255)
            
            smoked_images.append(result)
            smoke_masks.append(smoke)
    
    return smoked_images, smoke_masks

def process_single_image(img_file, clear_dir, output_dir, num_variations, use_gpu, seed_base):
    """Process a single image, generating smoky variations with a unique seed."""
    random.seed(seed_base)
    np.random.seed(seed_base)
    if use_gpu and torch is not None:
        torch.manual_seed(seed_base)
    try:
        img_path = os.path.join(clear_dir, img_file)
        img = cv2.imread(img_path)
        if img is None:
            logging.error(f"Failed to load {img_path}")
            return
        if img.shape[0] != 512 or img.shape[1] != 512:
            img = cv2.resize(img, (512, 512))
        base_name = os.path.splitext(img_file)[0]
        if use_gpu and torch is not None and torch.cuda.is_available():
            try:
                smoked_images, smoke_masks = add_smoke_to_image(img, use_gpu=True, num_variations=num_variations)
            except torch.cuda.OutOfMemoryError:
                logging.warning(f"GPU out of memory for {img_file}. Switching to CPU.")
                smoked_images, smoke_masks = add_smoke_to_image(img, use_gpu=False, num_variations=num_variations)
        else:
            smoked_images, smoke_masks = add_smoke_to_image(img, use_gpu=False, num_variations=num_variations)
        for i in range(num_variations):
            clear_path = os.path.join(output_dir, "clear", f"{base_name}_var{i}.png")
            blurry_path = os.path.join(output_dir, "blurry", f"{base_name}_var{i}.png")
            mask_path = os.path.join(output_dir, "masks", f"{base_name}_var{i}_mask.png")
            os.makedirs(os.path.dirname(clear_path), exist_ok=True)
            os.makedirs(os.path.dirname(blurry_path), exist_ok=True)
            os.makedirs(os.path.dirname(mask_path), exist_ok=True)
            cv2.imwrite(clear_path, img)
            cv2.imwrite(blurry_path, smoked_images[i])
            cv2.imwrite(mask_path, (smoke_masks[i] * 255).astype(np.uint8))
    except Exception as e:
        logging.error(f"Error processing {img_file}: {e}")

def process_dataset(clear_dir=CLEAR_DIR, output_dir=OUTPUT_DIR, num_variations=NUM_VARIATIONS, 
                   total_images=TOTAL_IMAGES, num_workers=NUM_WORKERS, use_gpu=USE_GPU, batch_size=100):
    """Process the entire dataset with multiprocessing, applying smoke effects."""
    os.makedirs(os.path.join(output_dir, "clear"), exist_ok=True)
    os.makedirs(os.path.join(output_dir, "blurry"), exist_ok=True)
    os.makedirs(os.path.join(output_dir, "masks"), exist_ok=True)
    
    image_files = [f for f in os.listdir(clear_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    if total_images is not None and total_images < len(image_files):
        image_files = image_files[:total_images]  # Take first N images instead of random sample
    
    total_to_process = len(image_files)
    logging.info(f"Processing {total_to_process} images with {num_variations} variations each...")
    
    # Process in batches to avoid memory issues and provide better progress feedback
    for batch_start in range(0, total_to_process, batch_size):
        batch_end = min(batch_start + batch_size, total_to_process)
        batch = image_files[batch_start:batch_end]
        
        logging.info(f"Processing batch {batch_start//batch_size + 1}/{(total_to_process-1)//batch_size + 1} ({len(batch)} images)")
        
        # Use a context manager to ensure proper cleanup
        with ProcessPoolExecutor(max_workers=num_workers) as executor:
            # Spread random seeds to avoid identical random patterns
            futures = []
            for i, img_file in enumerate(batch):
                seed = batch_start + i + 42  # Use a deterministic but varied seed
                futures.append(executor.submit(
                    process_single_image, img_file, clear_dir, output_dir, num_variations, use_gpu, seed
                ))
            
            # Process results as they complete and show progress
            completed = 0
            with tqdm(total=len(batch), desc=f"Batch {batch_start//batch_size + 1}") as pbar:
                for future in as_completed(futures):
                    try:
                        # Get result to catch any exceptions
                        future.result()
                    except Exception as e:
                        logging.error(f"Error in worker process: {e}")
                    finally:
                        completed += 1
                        pbar.update(1)
            
            logging.info(f"Completed {completed}/{len(batch)} images in current batch")
        
        # Force garbage collection between batches
        import gc
        gc.collect()
        if torch is not None and torch.cuda.is_available():
            torch.cuda.empty_cache()
            
        logging.info(f"Completed {batch_end}/{total_to_process} images total")

# === Optional Visualization Function ===

def visualize_sample(clear_img, smoky_img, mask, base_name, i):
    """Display sample images in Jupyter notebook using matplotlib."""
    import matplotlib.pyplot as plt
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.title("Clear")
    plt.imshow(cv2.cvtColor(clear_img, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.subplot(1, 3, 2)
    plt.title("Smoky")
    plt.imshow(cv2.cvtColor(smoky_img, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.subplot(1, 3, 3)
    plt.title("Mask")
    plt.imshow(mask, cmap="gray")
    plt.axis("off")
    plt.suptitle(f"{base_name}_var{i}")
    plt.show()

# === Main Execution Section ===

if __name__ == "__main__":
    # Sample 5 images first
    import matplotlib.pyplot as plt
    
    # Define function to process and display a single image
    def process_and_display_sample(img_path, idx):
        img = cv2.imread(img_path)
        if img is None:
            logging.error(f"Failed to load {img_path}")
            return
        
        # Resize if needed
        if img.shape[0] != 512 or img.shape[1] != 512:
            img = cv2.resize(img, (512, 512))
        
        # Set a fixed random seed for reproducibility
        seed = 42 + idx
        random.seed(seed)
        np.random.seed(seed)
        if torch is not None:
            torch.manual_seed(seed)
        
        # Generate smoke effect
        smoked_images, smoke_masks = add_smoke_to_image(img, num_variations=1)
        
        # Display results
        plt.figure(figsize=(15, 5))
        plt.subplot(1, 3, 1)
        plt.title("Clear")
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        
        plt.subplot(1, 3, 2)
        plt.title("Smoky")
        plt.imshow(cv2.cvtColor(smoked_images[0], cv2.COLOR_BGR2RGB))
        plt.axis("off")
        
        plt.subplot(1, 3, 3)
        plt.title("Mask")
        plt.imshow(smoke_masks[0], cmap="gray")
        plt.axis("off")
        
        plt.suptitle(f"Sample {idx+1}")
        plt.show()
        
        return smoked_images[0], smoke_masks[0]
    
    # Get image files
    image_files = [f for f in os.listdir(CLEAR_DIR) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    if len(image_files) == 0:
        logging.error(f"No image files found in {CLEAR_DIR}")
    else:
        # Select 5 random samples or fewer if not enough images
        num_samples = min(5, len(image_files))
        sample_files = random.sample(image_files, num_samples)
        
        print(f"Displaying {num_samples} sample images with smoke effect...")
        samples = []
        for i, file in enumerate(sample_files):
            img_path = os.path.join(CLEAR_DIR, file)
            smoked_img, smoke_mask = process_and_display_sample(img_path, i)
            samples.append((file, smoked_img, smoke_mask))
        
        # Ask for user input to continue
        user_input = input("Enter number of images to process (e.g., '100') or press Enter to process all: ")
        
        # Parse user input for potential limit
        num_to_process = None
        if user_input.strip() and user_input.strip().isdigit():
            num_to_process = int(user_input.strip())
            print(f"Will process {num_to_process} images.")
        else:
            print("Will process all images.")
    
        # Sequential processing only - no parallel processing
        print("Processing images sequentially...")
        to_process = image_files
        if num_to_process is not None:
            to_process = to_process[:num_to_process]
        
        # Prepare output directories
        os.makedirs(os.path.join(OUTPUT_DIR, "clear"), exist_ok=True)
        os.makedirs(os.path.join(OUTPUT_DIR, "blurry"), exist_ok=True)
        os.makedirs(os.path.join(OUTPUT_DIR, "masks"), exist_ok=True)
        
        # Process in batches for better progress tracking
        batch_size = 10
        total_batches = (len(to_process) + batch_size - 1) // batch_size
        
        for batch_idx in range(total_batches):
            start_idx = batch_idx * batch_size
            end_idx = min(start_idx + batch_size, len(to_process))
            batch = to_process[start_idx:end_idx]
            
            print(f"\nProcessing batch {batch_idx+1}/{total_batches} (images {start_idx+1}-{end_idx} of {len(to_process)})")
            
            for i, img_file in enumerate(tqdm(batch, desc=f"Batch {batch_idx+1}")):
                seed = start_idx + i
                try:
                    process_single_image(img_file, CLEAR_DIR, OUTPUT_DIR, NUM_VARIATIONS, USE_GPU, seed)
                except Exception as e:
                    logging.error(f"Error processing {img_file}: {e}")
                    continue
            
            # Force garbage collection between batches
            import gc
            gc.collect()
            if torch is not None and torch.cuda.is_available():
                torch.cuda.empty_cache()
        
        logging.info("Processing complete!")
```
